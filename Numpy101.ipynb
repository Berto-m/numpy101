### **The Dot Product**
image.png

````
a = np.array([1,2])
b = np.array([3,4])
````

````
dot = 0
for e,f in zip(a,b):
  dot = dot + (e*f)
dot
````

````
dot = 0
for i in range(len(a)):
  dot = dot + (a[i] * b[i])
dot
````

````
np.sum(a*b)
````

````
# a lot of these functions are instances methods you can call them on the numpy object as shown below
(a*b).sum()
````

````
# that is how to dot product manually
# but remember that numpy is was made to do only math
# that's why there is function for DOT PRODUCT
np.dot(a,b)
````
